```python
import pandas as pd
import numpy as np
from scipy.io.wavfile import write
from tqdm import tqdm
from time import sleep
import matplotlib.pyplot as plt
from scipy.signal import spectrogram

# Read the EEG data from a CSV file
data = pd.read_csv('eeg_data.csv', low_memory=False)  # Use low_memory=False to avoid dtype warnings
data['TimeStamp'] = pd.to_datetime(data['TimeStamp'])
data.set_index('TimeStamp', inplace=True)

# Sample 10% of the data continuously in time
total_rows = data.shape[0]
sample_size = int(0.10 * total_rows)
start_index = np.random.randint(0, total_rows - sample_size)
data_sample_continuous = data.iloc[start_index:start_index + sample_size].copy()

# Combine Delta columns into a single column for the sample
data_sample_continuous['Delta_combined'] = data_sample_continuous[['Delta_TP9', 'Delta_AF7', 'Delta_AF8', 'Delta_TP10']].sum(axis=1)

# Calculate the mean of Delta combined column in minute intervals for the sample
minute_resampled_sample = data_sample_continuous.index.to_period('Min')
data_sample_continuous['minute_resampled'] = data_sample_continuous.groupby(minute_resampled_sample)['Delta_combined'].transform('mean')

# Define frequencies for C major scale (do-re-mi-fa-sol-la-ti-do)
scale = np.array([261.63, 293.66, 329.63, 349.23, 392.00, 440.00, 493.88, 523.25])  # Frequencies in Hz

# Map EEG data continuously to musical notes in the C major scale for the sample
mapped_indices_sample = np.interp(data_sample_continuous['minute_resampled'], 
                                   (data_sample_continuous['minute_resampled'].min(), data_sample_continuous['minute_resampled'].max()), 
                                   (0, len(scale) - 1))  # Map to scale indices

# Convert indices to actual frequency values
mapped_frequencies_sample = scale[np.clip(np.round(mapped_indices_sample).astype(int), 0, len(scale) - 1)]

# Dynamic adjustment of pitch and volume factors based on wave shape for the sample
wave_amplitude_sample = np.abs(np.sin(mapped_frequencies_sample / max(scale)))  # Normalize wave shape
pitch_factor_sample = wave_amplitude_sample / np.max(wave_amplitude_sample)
volume_factor_sample = 0.5 + 0.5 * wave_amplitude_sample  # Adjust as needed for desired volume

# Plot the mapped frequencies with dynamically adjusted pitch and volume factors for the sample
plt.figure(figsize=(12, 6))
plt.scatter(data_sample_continuous.index, mapped_frequencies_sample, c=pitch_factor_sample, cmap='cool', s=10)
plt.colorbar(label='Pitch')
plt.scatter(data_sample_continuous.index, mapped_frequencies_sample, c=volume_factor_sample, cmap='viridis', s=10)
plt.colorbar(label='Volume')
plt.plot(data_sample_continuous.index, mapped_frequencies_sample, color='gray', linestyle='-', linewidth=0.5)
plt.title('Mapped Frequencies with Dynamically Adjusted Volume and Pitch (Sample)')
plt.xlabel('Time')
plt.ylabel('Frequency (Hz)')

# Save the plot as a PNG file
plt.savefig('mapped_frequencies_sample.png', format='png', dpi=300)  # Adjust dpi as needed
plt.close()  # Close the figure to free memory

# Generate pulse-wave sound data based on mapped frequencies of the sample
duration_sample = 30  # Duration in seconds for sample
sample_rate = 44100  # Sample rate for audio
time_values_sample = np.linspace(0, duration_sample, sample_rate * duration_sample)

print("Generating pulse-wave sound based on C major scale with volume and pitch adjustments for the sample...")
with tqdm(total=len(mapped_frequencies_sample)) as pbar:
    audio_data_sample = np.zeros(sample_rate * duration_sample)
    for i, freq in enumerate(mapped_frequencies_sample):
        audio_data_sample += volume_factor_sample[i] * np.sin(2 * np.pi * freq * time_values_sample)
        pbar.update(1)
        sleep(0.01)  # Simulate some processing time

# Normalize the audio data of the sample
audio_data_sample /= np.max(np.abs(audio_data_sample))

# Save the audio data of the sample as a WAV file
write('pulse_wave_sound_sample.wav', sample_rate, audio_data_sample.astype(np.float32))
print("\nSample musical audio file with volume and pitch adjustments generated successfully!")

# Generate spectrogram for the pulse-wave sound
frequencies_sample, times_sample, Sxx_sample = spectrogram(audio_data_sample, sample_rate)
plt.figure(figsize=(12, 6))
plt.pcolormesh(times_sample, frequencies_sample, 10 * np.log10(Sxx_sample), shading='gouraud')
plt.colorbar(label='Intensity (dB)')
plt.title('Spectrogram of Pulse-Wave Sound')
plt.ylabel('Frequency (Hz)')
plt.xlabel('Time (s)')
plt.savefig('spectrogram_pulse_wave_sound.png', format='png', dpi=300)
plt.close()

# Generate ECG-like beep sound based on sampled data
# Parameters for ECG-like sound
beep_duration = 0.1  # Duration of each beep in seconds
num_beeps = len(data_sample_continuous)  # Number of beeps based on the length of the sample
ecg_audio_data = np.zeros(int(sample_rate * num_beeps * (beep_duration + 0.9)))  # Total duration

# Create the ECG-like sound using peaks from the Delta_combined data
for i, value in enumerate(data_sample_continuous['Delta_combined']):
    if value > data_sample_continuous['Delta_combined'].mean():  # Only beep on significant peaks
        start_sample = int(i * (beep_duration + 0.9) * sample_rate)
        t = np.linspace(0, beep_duration, int(sample_rate * beep_duration), endpoint=False)
        ecg_audio_data[start_sample:start_sample + len(t)] += 0.5 * np.sin(2 * np.pi * 600 * t)

# Normalize the ECG audio data
ecg_audio_data /= np.max(np.abs(ecg_audio_data))

# Save the ECG-like beep sound as a WAV file
write('ecg_beep_sound.wav', sample_rate, ecg_audio_data.astype(np.float32))
print("\nECG-like beep audio file generated successfully!")

# Generate spectrogram for the ECG beep sound
frequencies_ecg, times_ecg, Sxx_ecg = spectrogram(ecg_audio_data, sample_rate)
plt.figure(figsize=(12, 6))
plt.pcolormesh(times_ecg, frequencies_ecg, 10 * np.log10(Sxx_ecg), shading='gouraud')
plt.colorbar(label='Intensity (dB)')
plt.title('Spectrogram of ECG Beep Sound')
plt.ylabel('Frequency (Hz)')
plt.xlabel('Time (s)')
plt.savefig('spectrogram_ecg_beep_sound.png', format='png', dpi=300)
plt.close()
```
